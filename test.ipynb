```python
from googleapiclient.discovery import build
import pandas as pd
from datetime import datetime,date
import time
import json
import os.path
from os import path
```

## API Calls
---

```python
api_key= input("API KEY ")
youtube = build("youtube","v3", developerKey=api_key)

date_new= date(datetime.now().year,datetime.now().month,datetime.now().day).isoformat()
hour=datetime.now().hour
title=f"{date_new[0:4]}_{date_new[5:7]}_{date_new[8:10]}_{datetime.now().hour}"

top_videos= []

chart_mx= youtube.videos().list(
part=["id","snippet","statistics"],
chart="mostPopular",
regionCode="MX",
maxResults=50).execute()
```

```python
print(json.dumps(chart_mx, indent = 4))
```

## Merging Files
---

```python
frames=[]

for i in range(hour):
    if path.exists(f"trending_videos_data/{title[0:11]}{i}.csv"): 
        frames.append(pd.read_csv(f"trending_videos_data/{title[0:11]}{i}.csv", index_col=0))

main_df = pd.concat(frames).drop_duplicates(subset=['video_title'], keep='last')
```
